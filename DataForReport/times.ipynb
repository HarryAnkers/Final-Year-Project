```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
filepath = './timelog.txt'
outputs = []
lines = []
tmplines = []
with open(filepath) as fp:
    for line in fp:
        if line == '-\n' or line == '-':
            lines.append(tmplines)
            tmplines=[]
        else:
            tmplines.append(float(line))
times = np.asarray(lines)

times_sorted = []
for i in range(6):
    times_sorted.append(times[:,i+6][~times[:,i].astype(bool)])

normal_times = np.divide(times[:,6:].T,times[:,9].T).T
```

```python
for i, op in enumerate(["op0","op1","op2","op3","op4","op5"]):
    print(op)
    print(np.mean(times_sorted[i]))
    print(np.std(times_sorted[i]))
    print(stats.skew(times_sorted[i]))
plt.boxplot(times_sorted, labels=["op0", "op1", "op2", "op3", "op3-ieee", "op3-fast",])
plt.xlabel("op. level")
plt.ylabel("times")
plt.title("done")
plt.grid()
plt.savefig('times-file-50bin.png', dpi=500)
plt.show()

print(np.mean(normal_times,axis=0))
```
